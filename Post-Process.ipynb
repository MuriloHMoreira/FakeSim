```python
import matplotlib.pyplot as plt
import numpy as np
import meshio
```

```python
reader = meshio.xdmf.TimeSeriesReader('./resultados/real_case_2D/Fields.xdmf')

vertex, cells_list = reader.read_points_cells()

cells_connectivity = cells_list[0].data
x_coords = vertex[:, 0]
y_coords = vertex[:, 1]
```

```python
i = 70
t_i = reader.read_data(i)[0]
point_fields_i = reader.read_data(i)[1]
T_i = point_fields_i['Temperature [K]'].flatten()
Pv_i = point_fields_i['Pressure [Pa]'].flatten()
```

```python
# %matplotlib notebook
plt.tricontourf(x_coords, y_coords, cells_connectivity, T_i - 273.15, levels=256, cmap='viridis')
plt.tricontourf(x_coords, y_coords, cells_connectivity, T_i - 273.15, levels=256, cmap='viridis')
plt.tricontourf(x_coords, y_coords, cells_connectivity, T_i - 273.15, levels=256, cmap='viridis')
plt.tricontourf(x_coords, y_coords, cells_connectivity, T_i - 273.15, levels=256, cmap='viridis')

plt.triplot(x_coords, y_coords, cells_connectivity, c='k', lw=0.3, alpha=0.2)

plt.gca().set_aspect('equal')

plt.colorbar(label='Temperatura [°C]')
plt.savefig('./teste.pdf')
plt.show()
```
